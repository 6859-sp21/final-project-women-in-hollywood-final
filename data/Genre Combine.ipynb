```python
import numpy as np
import pandas as pd
import requests
```

```python
bechdel = pd.read_csv("movies.csv")
bechdel = bechdel[['year', 'imdb', 'title', 'test', 'clean_test', 'binary', 'budget','domgross', 'intgross']]
bechdel.head()
```

```python
imdb = pd.read_csv("IMDb movies.csv")
imdb = imdb[['imdb_title_id', 'genre']]
imdb.columns = ['imdb', 'genre']
imdb.head()
```

```python
merged = bechdel.merge(imdb, on='imdb', how='left')
```

```python
merged.loc[merged['imdb']=='tt2243973', 'genre'] = 'Drama'
merged.loc[merged['imdb']=='tt1684628', 'genre'] = 'Drama'
merged.loc[merged['imdb']=='tt1720182', 'genre'] = 'Drama'
merged.loc[merged['imdb']=='tt1405500', 'genre'] = 'Drama'
merged.loc[merged['imdb']=='tt1274300', 'genre'] = "Drama, Fantasy"
merged.loc[merged['imdb']=='tt1232207', 'genre'] = "Documentary"
merged.loc[merged['imdb']=='tt0489018', 'genre'] = "Horror"
merged.loc[merged['imdb']=='tt1227929', 'genre'] = "Documentary"
merged.loc[merged['imdb']=='tt0368563', 'genre'] = "Thriller"
merged.loc[merged['imdb']=='tt0942903', 'genre'] = "Adventure, Sci-Fi"
merged.loc[merged['imdb']=='tt420238', 'genre'] = "Adventure, Family"
merged.loc[merged['imdb']=='tt1185616', 'genre'] = "Drama"
merged.loc[merged['imdb']=='tt00293564', 'genre'] = "Action, Comedy, Crime"
merged.loc[merged['imdb']=='tt0493430', 'genre'] = "Action, Comedy, Documentary"
merged.loc[merged['imdb']=='tt0474361', 'genre'] = "Documentary"
merged.loc[merged['imdb']=='tt0319970', 'genre'] = "Drama, Horror"
merged.loc[merged['imdb']=='tt0142032', 'genre'] = "Drama"
merged.loc[merged['imdb']=='tt0257001', 'genre'] = "Adventure, Fantasy, Sci-Fi, Family"
merged.loc[merged['imdb']=='tt0120108', 'genre'] = "Drama, Romance"
merged.loc[merged['imdb']=='tt0128996', 'genre'] = "Family, Fantasy, Romance"
merged.loc[merged['imdb']=='tt0112130', 'genre'] = "Drama"
```

```python
singles = set()
for i in merged['genre']:
    if str(i) == "nan":
        i = "BAD,"
    splitter = i.split(",")
    if len(splitter) == 1:
        singles.add(splitter[0])
print(singles)
# Genres: Action, Romance, Sci-Fi, Drama, Biography, Comedy, Fantasy, Horror, Thriller
```

```python
def genreCol(row, genre, genre2=None):
#     print(genre, row['genre'])
    if not genre2:
        if genre in row['genre']:
            return 1
        return 0
    if genre in row['genre'] or genre2 in row['genre']:
        return 1
    return 0

merged['Horror'] = merged.apply(lambda row: genreCol(row, 'Horror'), axis=1)
merged['Comedy'] = merged.apply(lambda row: genreCol(row, 'Comedy'), axis=1)
merged['Action'] = merged.apply(lambda row: genreCol(row, 'Action', 'Adventure'), axis=1)
merged['Romance'] = merged.apply(lambda row: genreCol(row, 'Romance'), axis=1)
merged['Sci-Fi'] = merged.apply(lambda row: genreCol(row, 'Sci-Fi'), axis=1)
merged['Drama'] = merged.apply(lambda row: genreCol(row, 'Drama'), axis=1)
merged['Historic Fiction'] = merged.apply(lambda row: genreCol(row, 'Biography', 'Documentary'), axis=1)
merged['Fantasy'] = merged.apply(lambda row: genreCol(row, 'Fantasy'), axis=1)
merged['Thriller'] = merged.apply(lambda row: genreCol(row, 'Thriller'), axis=1)
```

```python
merged['Sum'] = merged['Horror']+merged['Comedy']+merged['Action']+merged['Romance']+merged['Sci-Fi']+merged['Drama']+merged['Historic Fiction']+merged['Fantasy']+merged['Thriller']
merged[merged['Sum']<1]
```

```python
merged.to_csv("genres.csv", index=False)
```

```python
genres = pd.read_csv("genres.csv")
genres['helper'] = 0
genres.head()
```

```python
percentPassed = []
for i in ["Horror", "Comedy", "Action", "Romance", "Sci-Fi", "Drama", "Historic Fiction", "Fantasy", "Thriller"]:
    helper = genres.groupby(['year']).count()['helper']
    tot = helper + genres[genres[i]==1].groupby(['year']).count()['imdb']
    passed = genres[(genres[i]==1) & (genres['binary'] == "PASS")].groupby(['year']).count()['imdb']
    percentPassed.append([i] + list(passed/tot))
pPassed = pd.DataFrame(percentPassed, columns=['genre']+list(range(1973, 2014)))
pPassed.to_csv("genres2.csv", index=False)
```
